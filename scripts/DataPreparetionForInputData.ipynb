```python
import json
import os
import time
from openai import AzureOpenAI
import smtplib
from email.message import EmailMessage

# Initialize the environment
os.environ["OPENAI_API_TYPE"] = "azure"
os.environ["AZURE_OPENAI_ENDPOINT"] = "<your_azure_endpoint>"
os.environ["AZURE_OPENAI_KEY"] = "<your_azure_api_key>"
os.environ["OPENAI_API_VERSION"] = "2024-02-15-preview"

client = AzureOpenAI(
    azure_endpoint=os.getenv("AZURE_OPENAI_ENDPOINT"),
    api_key=os.getenv("AZURE_OPENAI_KEY"),
    api_version=os.getenv("OPENAI_API_VERSION")
)

print("Azure OpenAI client initialized successfully.")

def load_passages(directory_path):
    passages = []
    for filename in os.listdir(directory_path):
        if filename.endswith('.json'):
            file_path = os.path.join(directory_path, filename)
            with open(file_path, 'r') as file:
                data = json.load(file)
                if isinstance(data, list):
                    passages.extend(data)
                else:
                    passages.append(data)
    return passages

def ObligationCheck(passage):
    # Implement the real check here using a model or heuristic
    while True:
        try:
            messages = [
                {"role": "system", "content": "Review the following passage and determine if it contains any obligations. Return 'True' if there are obligations and 'False' otherwise:"},
                {"role": "user", "content": passage}
            ]

            # Make the API call
            chat_completion = client.chat.completions.create(
                messages=messages,
                model="gpt-4-turbo-1106",
                temperature=0.7,
                max_tokens=400,
                top_p=0.95,
                frequency_penalty=0,
                presence_penalty=0,
                stop=None
            )
            answ = chat_completion.choices[0].message.content.strip().lower()
            return answ == 'true'
        except Exception as e:
            print(f"An error occurred: {e}")
            print("Retrying in 20 seconds...")
            time.sleep(20)

def sendEmail(messagetext):
    sender = "Your Name <your_email@example.com>"
    receiver = "Recipient Name <recipient_email@example.com>"

    message = EmailMessage()
    message.set_content(f"{messagetext}\n")
    message['Subject'] = 'Code is Running'
    message['From'] = sender
    message['To'] = receiver

    try:
        with smtplib.SMTP("<your_smtp_server>", 587) as server:
            server.starttls()
            server.login("<your_smtp_user>", "<your_smtp_password>")
            server.send_message(message)
        print("Email sent successfully!")
    except Exception as e:
        print(f"Failed to send email: {e}")

def classify_passages(passages, topics_keywords):
    topic_to_passages = {topic: [] for topic in topics_keywords.keys()}
    last_email_time = time.time()

    for i, passage in enumerate(passages, start=1):
        text = passage['Passage']
        text_lower = text.lower()
        if '\\Table' in text or '\\Figure' in text or len(text.split()) > 200 or len(text.split()) < 20:
            continue

        for topic, keywords in topics_keywords.items():
            if any(keyword.lower() in text_lower for keyword in keywords):
                if ObligationCheck(text):
                    topic_to_passages[topic].append(passage)

        if time.time() - last_email_time >= 300:
            sendEmail(f"Processing update at {i} passages...")
            last_email_time = time.time()

    sendEmail("Code finished processing.")
    return topic_to_passages

# Paths setup
directory_path = 'FilteredDocuments'  # Update this path
output_path = 'topic_classifications.json'  # Update this path
topics_keywords = {}  # Provide your topics and keywords here

# Process the passages
passages = load_passages(directory_path)
topic_classifications = classify_passages(passages, topics_keywords)

# Save the results to a JSON file
with open(output_path, 'w') as f:
    json.dump(topic_classifications, f, indent=4)

print(f"Classification results saved to {output_path}")
```
